```python
import numpy as np
import matplotlib.pyplot as plt

# Signal parameters
T_total = 5  # Total duration in seconds
fs = 44100  # Sampling frequency in Hz

# Create time vector for the full duration
dt = 1 / fs
t_full = np.arange(0, T_total, dt)

# Frequencies to plot
f_values = [500, 5000, 10000]  # Frequencies in Hz

# Plot only a small segment to visualize the waveform clearly
plot_duration = 0.01  # seconds
t_plot = np.arange(0, plot_duration, dt)

# Generate and plot x(t) for each frequency
plt.figure(figsize=(15, 10))

for i, f in enumerate(f_values):
    x_t = np.cos(2 * np.pi * f * t_plot)

    plt.subplot(len(f_values), 1, i + 1)
    plt.plot(t_plot, x_t)
    plt.title(f'Sinal x(t) para f = {f} Hz (segmento de {plot_duration}s)')
    plt.xlabel('Tempo (s)')
    plt.ylabel('Amplitude')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from IPython.display import Audio

# Generate and play x(t) for each frequency for the full duration
print("Playing generated signals:")

for i, f in enumerate(f_values):
    x_t_full = np.cos(2 * np.pi * f * t_full)

    # Normalize the signal to ensure it's within the audible range (-1 to 1)
    x_t_full_normalized = x_t_full / np.max(np.abs(x_t_full))

    print(f'Sinal x(t) para f = {f} Hz:')
    display(Audio(x_t_full_normalized, rate=fs))
```
